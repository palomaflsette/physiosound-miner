```python
import sys
import os
sys.path.append(os.path.abspath(".."))
import numpy as np
import matplotlib.pyplot as plt
from preprocessing.features.takens_rqa import extract_rqa_features
```

```python
fs = 1000  # Hz
t = np.linspace(0, 2, 2 * fs, endpoint=False)
signal = np.sin(2 * np.pi * 5 * t) + 0.3 * np.random.randn(len(t))

plt.figure(figsize=(10, 3))
plt.plot(t, signal)
plt.title("Sinal Simulado (senide + rudo)")
plt.xlabel("Tempo [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
rqa_results = extract_rqa_features(
    signal=signal,
    fs=fs,
    file_id="sinal_simulado",
    window_id=0,
    embedding_dimension=3,
    time_delay=2,
    threshold=0.5,
    normalize=True,
    verbose=True
)
```
